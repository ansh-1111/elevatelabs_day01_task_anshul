```python
import pandas as pd

# Use the raw URL
df = pd.read_csv("https://raw.githubusercontent.com/ansh-1111/elevatelabs_day01_task_anshul/main/Data/Mall_Customers.csv")
df.head()
```

```python
# Check number of rows and columns
df.shape

# Check column names
df.columns

# Check data types and null values
df.info()

# Quick stats for numeric columns
df.describe()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={
    "Annual Income (k$)": "Annual_Income",
    "Spending Score (1-100)": "Spending_Score"
}, inplace=True)
```

```python
df['Gender'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Age distribution
sns.histplot(df['Age'], bins=10)
plt.show()

# Annual Income vs Spending Score
sns.scatterplot(data=df, x='Annual_Income', y='Spending_Score', hue='Gender')
plt.show()
```

```python
df.to_csv('Mall_Customers_cleaned.csv', index=False)
```
